```python
ratings=merged_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# OBSERVATION
- WHEN RATING IS BETWEEN FROM 4.5 TO 4.9 THE RATING WERE EXCELLENT

- WHEN RATING IS BETWEEN FROM 4.0 TO 4.4 THE RATING WERE VERY GOOD

- WHEN RATING IS BETWEEN FROM 3.5 TO 3.9 THE RATING WERE GOOD

- WHEN RATING IS BETWEEN FROM 2.5 TO 3.4 THE RATING WERE AVERAGE

- WHEN RATING IS BETWEEN FROM 1.8 TO 2.4 THE RATING WERE POOR

- WHEN RATING IS 0 IT MEANS PEOPLE HAS NOT GIVEN ANY RATING

```python
plt.figure(figsize=(12,7),dpi=300)
sns.set_style('whitegrid')
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings);
```

```python
plt.figure(figsize=(14,7),dpi=300)
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
plt.legend(loc=1);
```

#### OBSERVATION
- NOT RATED COUNT IS VERY HIGH

- MAXIMUM NUMBER OF RATINGS ARE BETWEEN 2.5 TO 3.4
